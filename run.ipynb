```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

```python
df=pd.DataFrame(pd.read_csv("Creditcard_data.csv"))
df.head()
```

```python
df=df.drop(["Time"],axis=1)
```

```python
df['Amount'] = StandardScaler().fit_transform(np.array(df['Amount']).reshape(-1, 1))
X=df.iloc[:,:29]
y=df['Class']
```

```python
sm = SMOTE(random_state = 42)
X_train_res, y_train_res = sm.fit_resample(X,y)
```

```python
print(X_train_res)
```

```python
df_new=pd.concat([pd.DataFrame(X_train_res),pd.DataFrame(y_train_res)],axis=1)
df_new
X=df_new.iloc[:,:29]
y=df_new['Class']
df_new
```

```python
n=1.96*1.96*0.5*0.5/(0.05*0.05)
print(n)
```

```python
sample_df = df_new.sample(int(n))
sample_df
```

```python
x=sample_df.iloc[:,:29]
y=sample_df['Class']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=1)

sample_accuracy=[]

GNB = GaussianNB()
GNB.fit(X_train, y_train)
y_pred = GNB.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
sample_accuracy.append(accuracy_score(y_test, y_pred)*100)

clf = RandomForestClassifier(n_estimators = 100)  
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
sample_accuracy.append(accuracy_score(y_test, y_pred)*100)

xg = xgb.XGBClassifier()
xg.fit(X_train, y_train)
y_pred = xg.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
sample_accuracy.append(accuracy_score(y_test, y_pred)*100)

adb = AdaBoostClassifier()
adb_model = adb.fit(X_train,y_train)
y_pred = adb_model.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
sample_accuracy.append(accuracy_score(y_test, y_pred)*100)

classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
sample_accuracy.append(accuracy_score(y_test, y_pred)*100)
```

```python
def Cluster_Sampling(df, cluster_size, n_clusters):
    N = len(df)
    K = int(N/cluster_size)
    data = None
    for k in range(K):
        sample_k = df.sample(cluster_size)
        sample_k["cluster"] = np.repeat(k,len(sample_k))
        df = df.drop(index = sample_k.index)
        data = pd.concat([data,sample_k],axis = 0)
    random_chosen_clusters = np.random.randint(0,K,size =n_clusters )
    samples = data[data.cluster.isin(random_chosen_clusters)]
    return(samples)
cluster_df = Cluster_Sampling(df = df_new, cluster_size = 20, n_clusters = 28)
```

```python
x=cluster_df.iloc[:,:29]
y=cluster_df['Class']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=1)

cluster_accuracy=[]

GNB = GaussianNB()
GNB.fit(X_train, y_train)
y_pred = GNB.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
cluster_accuracy.append(accuracy_score(y_test, y_pred)*100)

clf = RandomForestClassifier(n_estimators = 100)  
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
cluster_accuracy.append(accuracy_score(y_test, y_pred)*100)

xg = xgb.XGBClassifier()
xg.fit(X_train, y_train)
y_pred = xg.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
cluster_accuracy.append(accuracy_score(y_test, y_pred)*100)

adb = AdaBoostClassifier()
adb_model = adb.fit(X_train,y_train)
y_pred = adb_model.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
cluster_accuracy.append(accuracy_score(y_test, y_pred)*100)

classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
cluster_accuracy.append(accuracy_score(y_test, y_pred)*100)
```

```python
def systematic_sampling(df, step):
 
    indexes = np.arange(0, len(df), step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample

systematic_df = systematic_sampling(df_new, 5)
```
